# Описательные статистики числовых переменных

## Определение статистических показателей

### Базовые показатели распределения

**count** (количество наблюдений)
- Определение: Число непустых значений в переменной
- Интерпретация: Объем доступных данных для анализа
- Значение: Показывает полноту данных по каждой переменной

**mean** (среднее арифметическое)
- Определение: Сумма всех значений, деленная на их количество
- Формула: mean = Σx_i / n
- Интерпретация: Центр распределения данных
- Особенности: Чувствителен к выбросам

**median** (медиана)
- Определение: Значение, которое делит упорядоченный набор данных пополам
- Интерпретация: Устойчивая мера центра распределения
- Преимущество: Не чувствительна к выбросам

### Показатели вариабельности

**range** (размах)
- Определение: Разность между максимальным и минимальным значениями
- Формула: range = max - min
- Интерпретация: Общий разброс данных
- Ограничение: Сильно зависит от выбросов

**variance** (дисперсия)
- Определение: Средний квадрат отклонений от среднего значения (или разброс значений случайных величин в районе мат ожидания)
- Формула: variance = Σ(x_i - mean)² / (n-1)
- Интерпретация: Мера разброса данных вокруг среднего
- Особенности: Используется несмещенная оценка (ddof=1)

**std_dev** (стандартное отклонение)
- Определение: Квадратный корень из дисперсии
- Формула: std_dev = √variance
- Интерпретация: Среднее расстояние от точек данных до среднего
- Преимущество: Выражается в исходных единицах измерения

### Квантили распределения

**q_0.01** (1-й процентиль)
- Определение: Значение, ниже которого находится 1% данных
- Интерпретация: Нижняя граница распределения
- Применение: Выявление экстремально низких значений

**q_0.05** (5-й процентиль)
- Определение: Значение, ниже которого находится 5% данных
- Интерпретация: Нижний квантиль распределения
- Применение: Анализ левого хвоста распределения

**q_0.95** (95-й процентиль)
- Определение: Значение, ниже которого находится 95% данных
- Интерпретация: Верхний квантиль распределения
- Применение: Анализ правого хвоста распределения

**q_0.99** (99-й процентиль)
- Определение: Значение, ниже которого находится 99% данных
- Интерпретация: Верхняя граница распределения
- Применение: Выявление экстремально высоких значений

````python
import pandas as pd
import numpy as np

# Создаем функцию для расчета всех характеристик
def calculate_descriptive_stats(df, numeric_columns):
    stats_dict = {}
    
    for col in numeric_columns:
        data = df[col].dropna()  # Убираем пропущенные значения
        
        if len(data) > 0:
            stats_dict[col] = {
                'count': len(data),
                'min': np.min(data),
                'max': np.max(data),
                'range': np.max(data) - np.min(data),
                'mean': np.mean(data),
                'variance': np.var(data, ddof=1),  # несмещенная дисперсия
                'std_dev': np.std(data, ddof=1),   # несмещенное стандартное отклонение
                'median': np.median(data),
                'q_0.01': np.quantile(data, 0.01),
                'q_0.05': np.quantile(data, 0.05),
                'q_0.95': np.quantile(data, 0.95),
                'q_0.99': np.quantile(data, 0.99)
            }
    
    return pd.DataFrame(stats_dict).T

# Убедимся, что cleaned_file существует
if 'cleaned_file' not in dir():
    print("Ошибка: переменная cleaned_file не определена!")
    # Здесь нужно создать или загрузить наш DataFrame
    # cleaned_file = наш способ создания dataframe()
else:
    # Определяем numeric_columns если нужно
    if 'numeric_columns' not in dir():
        numeric_columns = cleaned_file.select_dtypes(include=[np.number]).columns.tolist()
    
    # Расчет статистик
    descriptive_stats = calculate_descriptive_stats(cleaned_file, numeric_columns)
    
    # Отображаем результаты
    print("ОПИСАТЕЛЬНЫЕ СТАТИСТИКИ ДЛЯ ВСЕХ ЧИСЛОВЫХ ПЕРЕМЕННЫХ")
    print("=" * 100)
    print(descriptive_stats)
````

Гистограммы предоставляют визуальное представление распределения числовых переменных, позволяя оценить форму распределения, центральную тенденцию, разброс и наличие выбросов.

### Структура гистограммы

**Столбцы (bins)**:

- Количество: 30 интервалов для детального анализа распределения

**Оси графика**:
- Ось X (Значение): Диапазон значений переменной
- Ось Y (Частота): Количество наблюдений в каждом интервале

**Заголовок подграфика**:
- Название переменной
- Среднее значение (mean) с точностью до двух знаков после запятой
### Параметры визуализации

**Прозрачность (alpha=0.7)**:
- Назначение: Улучшение читаемости при наложении нескольких графиков
- Эффект: Позволяет видеть плотность распределения

**Размер сетки графиков**:
- Адаптивная сетка: автоматическое определение строк и столбцов
- Максимум: 4 столбца для оптимального использования пространства
- Высота: пропорциональна количеству строк (5*n_rows)

## Интерпретация распределений

**Нормальное распределение**:
- Характеристики: Симметричная колоколообразная форма
- Признаки: mean ≈ median, умеренный разброс

**Скошенное распределение**:
- Правостороннее: Длинный хвост вправо, mean > median
- Левостороннее: Длинный хвост влево, mean < median
**Бимодальное распределение**:
- Признаки: Два выраженных пика
- Интерпретация: Возможное наличие двух различных групп в данных

### Анализ центра распределения

**Среднее значение (mean)**:
- Показатель в заголовке: Ориентир для оценки центра распределения
- Сравнение с медианой: Оценка симметрии распределения
- Интерпретация: Учет чувствительности к выбросам

### Выявление аномалий

**Выбросы**:
- Визуальные признаки: Изолированные столбцы на краях распределения
- Количественная оценка: Сравнение с квантилями q_0.01 и q_0.99

**Пропуски в данных**:
- Признаки: Неожиданные провалы в распределении
- Интерпретация: Возможные проблемы с сбором данных

# Визуализация распределения с помощью Boxplot

## Назначение диаграмм размаха (Boxplot)

Boxplot (ящик с усами) - это статистическая визуализация, которая предоставляет компактное представление о распределении числовых переменных, позволяя оценить центральную тенденцию, разброс, асимметрию и наличие выбросов.

## Структура Boxplot

### Основные элементы диаграммы

**Центральный ящик (Box)**:
- Представляет межквартильный размах (IQR) - средние 50% данных
- Нижняя граница: первый квартиль (Q1 - 25-й процентиль)
- Верхняя граница: третий квартиль (Q3 - 75-й процентиль)
- Линия внутри ящика: медиана (Q2 - 50-й процентиль)

**Усы (Whiskers)**:
- Верхний ус: extends to Q3 + 1.5 × IQR или до максимального значения в этом диапазоне
- Нижний ус: extends to Q1 - 1.5 × IQR или до минимального значения в этом диапазоне
- Длина усов показывает разброс "нормальных" данных

**Выбросы (Outliers)**:
- Точки за пределами усов
- Определяются как значения за пределами Q1 - 1.5×IQR и Q3 + 1.5×IQR

## Интерпретация Boxplot

### Анализ центра распределения

**Положение медианы**:
- Центр ящика: симметричное распределение
- Смещение вверх: правосторонняя асимметрия
- Смещение вниз: левосторонняя асимметрия

**Сравнение с гистограммами**:
- Boxplot дополняет гистограммы компактным представлением
- Позволяет быстро сравнить несколько распределений

### Оценка разброса данных

**Длина ящика (IQR)**:
- Короткий ящик: низкая изменчивость в центральной части
- Длинный ящик: высокая изменчивость в центральной части

**Длина усов**:
- Короткие усы: данные сконцентрированы вокруг центра
- Длинные усы: значительный разброс крайних значений

### Выявление выбросов

**Интерпретация выбросов**:
- Единичные точки: возможные ошибки измерения
- Группы точек: реальные экстремальные значения
- Симметричные выбросы: особенности распределения

**Метод определения**:
- Нижние выбросы: < Q1 - 1.5 × IQR
- Верхние выбросы: > Q3 + 1.5 × IQR

# Анализ распределения через сравнение квантилей

## Методология анализа распределения

### Концепция квантильного анализа

Квантильный анализ представляет собой мощный метод исследования распределения данных, основанный на сравнении различных процентилей распределения. Этот подход позволяет количественно оценить симметрию, разброс и форму распределения без предположений о его параметрической форме.
### Интерпретация коэффициента симметрии

**Симметричное распределение**:
- Коэффициент ≈ 1.0
- Диапазон: 0.67 - 1.5
- Характеристика: Равномерное распределение данных относительно медианы

**Правосторонняя асимметрия**:
- Коэффициент > 1.5
- Характеристика: Длинный правый хвост распределения
- Признак: Наличие экстремально высоких значений

**Левосторонняя асимметрия**:
- Коэффициент < 0.67
- Характеристика: Длинный левый хвост распределения
- Признак: Наличие экстремально низких значений

## Практическая интерпретация результатов

### Для симметричного распределения
- Коэффициент симметрии в диапазоне 0.67-1.5
- Распределение близко к нормальному
- Подходит для параметрических статистических тестов

### Для асимметричных распределений

**Правосторонняя асимметрия**:
- Преобладание более высоких значений
- Возможность логарифмического преобразования
- Рассмотрение robust статистических методов

**Левосторонняя асимметрия**:
- Преобладание более низких значений
- Возможность квадратичного преобразования
- Анализ причин смещения распределения
    

````python
# Анализ распределения через сравнение квантилей
print("АНАЛИЗ РАСПРЕДЕЛЕНИЯ ЧЕРЕЗ КВАНТИЛИ")
print("=" * 100)

for col in numeric_columns:
    if col in descriptive_stats.index:
        stats = descriptive_stats.loc[col]
        
        # Анализ симметрии распределения
        q1_to_median = stats['median'] - stats['q_0.05']
        median_to_q3 = stats['q_0.95'] - stats['median']
        symmetry_ratio = median_to_q3 / q1_to_median if q1_to_median != 0 else np.inf

        print(f"\n|-| {col}:")
        print(f"   Интервал 90% данных: [{stats['q_0.05']:.4f}, {stats['q_0.95']:.4f}]")
        print(f"   Интервал 98% данных: [{stats['q_0.01']:.4f}, {stats['q_0.99']:.4f}]")
        print(f"   Отношение (Q95-median)/(median-Q05): {symmetry_ratio:.4f}")
        
        if symmetry_ratio > 1.5:
            print("   !  Распределение скошено вправо")
        elif symmetry_ratio < 0.67:
            print("   !  Распределение скошено влево")
        else:
            print("   + Распределение относительно симметрично")
````

# Анализ корреляционной матрицы

## Методология анализа корреляций

### Тепловая карта корреляций

Тепловая карта корреляций представляет собой визуальное отображение матрицы попарных корреляций между числовыми переменными. Этот метод позволяет быстро оценить структуру взаимосвязей в данных и выявить группы сильно связанных переменных.

### Параметры визуализации

**Цветовая схема (cmap='coolwarm')**:
- Синие оттенки: отрицательные корреляции
- Белый/светлые цвета: корреляции близкие к нулю
- Красные оттенки: положительные корреляции

**Маска верхнего треугольника (mask)**:
- Исключает дублирующую информацию
- Упрощает чтение матрицы
- Показывает только уникальные пары корреляций

**Аннотации (annot=True)**:
- Числовые значения коэффициентов корреляции
- Формат отображения: два знака после запятой (fmt='.2f')
- Позволяет точное чтение значений корреляций

## Интерпретация коэффициентов корреляции

### Шкала силы корреляций

**Сильные корреляции**:
- 0.7 - 1.0: Очень сильная положительная связь
- -1.0 - -0.7: Очень сильная отрицательная связь

**Умеренные корреляции**:
- 0.3 - 0.7: Умеренная положительная связь
- -0.7 - -0.3: Умеренная отрицательная связь

**Слабые корреляции**:
- 0.0 - 0.3: Слабая положительная связь
- -0.3 - 0.0: Слабая отрицательная связь

### Интерпретация высоких корреляций

**Положительные корреляции > 0.7**:
- Переменные измеряют сходные конструкты
- Возможна дублирующая информация
- Кандидаты для объединения в факторы

**Отрицательные корреляции < -0.7**:
- Переменные измеряют противоположные аспекты
- Могут формировать биполярные факторы
- Важны для содержательной интерпретации

````python
# Тепловая карта корреляций
if len(numeric_columns) > 1:
    plt.figure(figsize=(12, 10))
    correlation_matrix = cleaned_file[numeric_columns].corr()
    
    mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))
    sns.heatmap(correlation_matrix, 
                mask=mask,
                annot=True, 
                cmap='coolwarm', 
                center=0,
                square=True,
                fmt='.2f',
                cbar_kws={'shrink': 0.8})
    
    plt.title('Матрица корреляций числовых переменных', fontsize=16, pad=20)
    plt.tight_layout()
    plt.show()
    
    # Поиск высоких корреляций
    high_corr = []
    for i in range(len(correlation_matrix.columns)):
        for j in range(i+1, len(correlation_matrix.columns)):
            if abs(correlation_matrix.iloc[i, j]) > 0.7:
                high_corr.append((
                    correlation_matrix.columns[i],
                    correlation_matrix.columns[j],
                    correlation_matrix.iloc[i, j]
                ))
    
    if high_corr:
        print("\nВысокие корреляции (> 0.7):")
        for corr in high_corr:
            print(f"{corr[0]} - {corr[1]}: {corr[2]:.3f}")
````
