# –î–ò–°–ö–†–ò–ú–ò–ù–ê–ù–¢–ù–´–ô –ê–ù–ê–õ–ò–ó –î–ê–ù–ù–´–•

# –ù–∞—á–∞–ª—å–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report

df_clust = pd.read_excel("Annual_Clasters.xlsx")
df_clust
```

```python
X = df_clust.copy()
X = X.drop(columns=[X.columns[-1]])
X = X.drop(columns=[X.columns[-1]])
y = df_clust["–ö–ª–∞—Å—Ç–µ—Ä"]
```

# –ü–µ—Ä–µ–∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏—è –æ–±—É—á–∞—é—â–µ–π –≤—ã–±–æ—Ä–∫–∏

```python
lda_full = LinearDiscriminantAnalysis()
lda_full.fit(X, y)
y_pred_full = lda_full.predict(X)
cm_full = confusion_matrix(y, y_pred_full)

header = "| –ß–∞—Å—Ç–æ—Ç–∞ | –ù–æ–º–µ—Ä –∫–ª–∞—Å—Å–∞ | –ü—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–Ω–∞—è –ø—Ä–∏–Ω–∞–¥–ª–µ–∂–Ω–æ—Å—Ç—å –∫ –∫–ª–∞—Å—Å—É | –í—Å–µ–≥–æ |"
print(header)
print("|" + "-"*(len(header)-2) + "|")
print("|         | (—Ä–µ–π—Ç–∏–Ω–≥)    |   1   |   2   |   3   |   4   | –Ω–∞–±–ª—é–¥–µ–Ω–∏–π |")
print("|" + "-"*(len(header)-2) + "|")

for i in range(4):
    row = f"| –ß–∞—Å—Ç–æ—Ç–∞ | {i+1:^12} |"
    for j in range(4):
        row += f" {cm_full[i,j]:^6} |"
    row += f" {cm_full[i].sum():^10} |"
    print(row)
print("|" + "-"*(len(header)-2) + "|")

for i in range(4):
    row = f"|    %    | {i+1:^12} |"
    for j in range(4):
        percentage = (cm_full[i,j] / cm_full[i].sum()) * 100
        row += f" {percentage:>6.1f} |"
    row += f" {100:>10.1f} |"
    print(row)
print("|" + "-"*(len(header)-2) + "|")
```

# –†–∞—Å—á–µ—Ç –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–µ–π –æ—à–∏–±–æ–∫ –ø—Ä–∏ –ø–µ—Ä–µ–∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏ –æ–±—É—á–∞—é—â–µ–π –≤—ã–±–æ—Ä–∫–∏

```python
print("| –ö–ª–∞—Å—Å (—Ä–µ–π—Ç–∏–Ω–≥) k | –û—Ü–µ–Ω–∫–∞ –∞–ø—Ä–∏–æ—Ä–Ω–æ–π –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ –∫–ª–∞—Å—Å–∞ œÄk | –û—Ü–µ–Ω–∫–∞ —É—Å–ª–æ–≤–Ω–æ–π –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ –æ—à–∏–±–∫–∏ Pk |")
print("|" + "-"*70 + "|")

class_probs = np.bincount(y) / len(y)
class_errors = []
for i in range(4):
    correct = cm_full[i,i]
    total = cm_full[i].sum()
    error = 1 - (correct / total) if total > 0 else 0
    class_errors.append(error)
    print(f"| {i+1:^18} | {class_probs[i]:>37.3f} | {error:>36.3f} |")
print("|" + "-"*70 + "|")

unconditional_error = 1 - np.trace(cm_full) / cm_full.sum()
print(f"| –ë–µ–∑—É—Å–ª–æ–≤–Ω–∞—è –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –æ—à–∏–±–∫–∏ P | {unconditional_error:>41.3f} ({unconditional_error*100:.1f}%) |")
print("|" + "-"*70 + "|")
```

# –í—ã–≤–æ–¥ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–æ–≤ –∫–ª–∞—Å—Å–∏—Ñ–∏—Ü–∏—Ä—É—é—â–∏—Ö —Ñ—É–Ω–∫—Ü–∏–π

```python
coef_full = lda_full.coef_
intercept_full = lda_full.intercept_
coefficient_columns = X.columns
coef_df = pd.DataFrame()
coef_df['–ü—Ä–∏–∑–Ω–∞–∫'] = list(coefficient_columns) + ['–ö–æ–Ω—Å—Ç–∞–Ω—Ç–∞']

for i in range(4):
    coef_values = list(coef_full[i]) + [intercept_full[i]]
    coef_df[f'd_{i+1}'] = coef_values

print(coef_df.to_string(index=False))
```

# –°–æ–∑–¥–∞–Ω–∏–µ —ç–∫–∑–∞–º–µ–Ω–∞—Ü–∏–æ–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–∏

```python
from sklearn.model_selection import train_test_split

if hasattr(X, 'iloc'):
    X_array = X.values
else:
    X_array = X
X_train, X_test, y_train, y_test = train_test_split(
    X_array, y, test_size=0.1, stratify=y, random_state=42
)

print(f"\n–†–∞–∑–º–µ—Ä –æ–±—É—á–∞—é—â–µ–π –≤—ã–±–æ—Ä–∫–∏: {len(X_train)} –Ω–∞–±–ª—é–¥–µ–Ω–∏–π")
print(f"–†–∞–∑–º–µ—Ä —ç–∫–∑–∞–º–µ–Ω–∞—Ü–∏–æ–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–∏: {len(X_test)} –Ω–∞–±–ª—é–¥–µ–Ω–∏–π")
print(f"–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø–æ –∫–ª–∞—Å—Å–∞–º –≤ –æ–±—É—á–∞—é—â–µ–π –≤—ã–±–æ—Ä–∫–µ: {np.bincount(y_train, minlength=4)}")
print(f"–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø–æ –∫–ª–∞—Å—Å–∞–º –≤ —ç–∫–∑–∞–º–µ–Ω–∞—Ü–∏–æ–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ: {np.bincount(y_test, minlength=4)}")
print("\n–ü—Ä–æ–≤–µ—Ä–∫–∞ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–Ω–æ—Å—Ç–∏ –∫–ª–∞—Å—Å–æ–≤:")
for i in range(1, 5):
    train_count = np.sum(y_train == i)
    test_count = np.sum(y_test == i)
    print(f"–ö–ª–∞—Å—Å {i+1}: –æ–±—É—á–∞—é—â–∞—è - {train_count}, —ç–∫–∑–∞–º–µ–Ω–∞—Ü–∏–æ–Ω–Ω–∞—è - {test_count}")
```

# –†–µ–∑—É–ª—å—Ç–∞—Ç—ã –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏ —ç–∫–∑–∞–º–µ–Ω–∞—Ü–∏–æ–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–∏

```python
lda_train = LinearDiscriminantAnalysis()
lda_train.fit(X_train, y_train)
y_test_pred = lda_train.predict(X_test)
cm_test = confusion_matrix(y_test, y_test_pred)

header = "|          |–ù–æ–º–µ—Ä –∫–ª–∞—Å—Å–∞|–ü—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–Ω–∞—è –ø—Ä–∏–Ω–∞–¥–ª–µ–∂–Ω–æ—Å—Ç—å –∫ –≥—Ä—É–ø–ø–µ    |–ò—Ç–æ–≥–æ|"
print(header)
print("|" + "-"*(len(header)-2) + "|")
print("|          | (—Ä–µ–π—Ç–∏–Ω–≥)  |  1  |  2  |  3  |  4  |          |")
print("|" + "-"*(len(header)-2) + "|")

for i in range(4):
    row = f"| –ß–∞—Å—Ç–æ—Ç–∞  | {i+1:<10} |"
    for j in range(4):
        row += f" {cm_test[i,j]:>3} |"
    row += f" {cm_test[i].sum():>8} |"
    print(row)

print("|" + "-"*(len(header)-2) + "|")

for i in range(4):
    row = f"|    %     | {i+1:<10} |"
    for j in range(4):
        percentage = (cm_test[i,j] / cm_test[i].sum()) * 100 if cm_test[i].sum() > 0 else 0
        row += f" {percentage:>5.1f} |"
    row += f" {100:>8.1f} |"
    print(row)

print("|" + "-"*(len(header)-2) + "|")

test_counts = np.bincount(y_test, minlength=4)
test_total = len(y_test)
row = "|–¢–µ—Å—Ç–æ–≤–∞—è  |            |"
for count in test_counts:
    percentage = (count / test_total) * 100
    row += f" {percentage:>5.1f} |"
row += f" {test_total:>8} |"
print(row)
print("|–≤—ã–±–æ—Ä–∫–∞   |            |" + "       |"*4 + "          |")
print("|" + "-"*(len(header)-2) + "|")
```

# –†–∞—Å—á–µ—Ç –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–µ–π –æ—à–∏–±–æ–∫ –¥–ª—è —ç–∫–∑–∞–º–µ–Ω–∞—Ü–∏–æ–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–∏

```python
print("| –ö–ª–∞—Å—Å (—Ä–µ–π—Ç–∏–Ω–≥) k | –û—Ü–µ–Ω–∫–∞ –∞–ø—Ä–∏–æ—Ä–Ω–æ–π –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ –∫–ª–∞—Å—Å–∞ Tk | –û—Ü–µ–Ω–∫–∞ —É—Å–ª–æ–≤–Ω–æ–π –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ –æ—à–∏–±–∫–∏ Pk |")
print("|" + "-"*70 + "|")
class_probs = np.bincount(y) / len(y)

test_errors = []
for i in range(4):
    if cm_test[i].sum() > 0:
        correct = cm_test[i,i]
        error = 1 - (correct / cm_test[i].sum())
    else:
        error = 0
    test_errors.append(error)
    print(f"| {i+1:^18} | {class_probs[i]:>37.3f} | {error:>36.3f} |")

print("|" + "-"*70 + "|")
unconditional_error_test = 1 - np.trace(cm_test) / cm_test.sum()
print(f"| –ë–µ–∑—É—Å–ª–æ–≤–Ω–∞—è –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –æ—à–∏–±–∫–∏ P | {unconditional_error_test:>41.3f} ({unconditional_error_test*100:.1f}%) |")
print("|" + "-"*70 + "|")
```

# –í—ã–≤–æ–¥ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–æ–≤ –∫–ª–∞—Å—Å–∏—Ñ–∏—Ü–∏—Ä—É—é—â–∏—Ö —Ñ—É–Ω–∫—Ü–∏–π

```python
coef_train = lda_train.coef_
intercept_train = lda_train.intercept_

coef_train_df = pd.DataFrame()
coef_train_df['–ü—Ä–∏–∑–Ω–∞–∫'] = list(coefficient_columns) + ['–ö–æ–Ω—Å—Ç–∞–Ω—Ç–∞']

for i in range(4):
    coef_values = list(coef_train[i]) + [intercept_train[i]]
    coef_train_df[f'd_{i+1}'] = coef_values

print(coef_train_df.to_string(index=False))
```

# –ê–Ω–∞–ª–∏–∑ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–π –ø—Ä–æ–≥–Ω–æ–∑–Ω—ã—Ö —Ä–µ–π—Ç–∏–Ω–≥–æ–≤

```python
deviations_full = y_pred_full - y
deviations_test = y_test_pred - y_test

def calculate_deviation_stats(deviations):
    total = len(deviations)
    stats = {}
    for k in range(-3, 4):
        count = np.sum(deviations == k)
        stats[k] = (count / total) * 100
    within_one_class = np.sum(np.abs(deviations) <= 1) / total * 100
    return stats, within_one_class

stats_full, within_one_full = calculate_deviation_stats(deviations_full)
stats_test, within_one_test = calculate_deviation_stats(deviations_test)

header = "| –ê–ª–≥–æ—Ä–∏—Ç–º               | –î–æ–ª—è –Ω–µ—Å–æ–≤–ø–∞–¥–µ–Ω–∏–π —Å –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ–º –Ω–∞ k –∫–ª–∞—Å—Å–æ–≤, –≤ %           | % –ø—Ä–∏ |k|‚â§1 |"
print(header)
subheader = "|" + " " * 23 + "|" + "".join([f" {k:>3} |" for k in range(-3, 4)]) + "          |"
print(subheader)
print("|" + "-"*84 + "|")

row_full = "| –†–µ–∂–∏–º –ø–µ—Ä–µ–∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏ |"
for k in range(-3, 4):
    row_full += f" {stats_full[k]:>5.2f} |"
row_full += f" {within_one_full:>9.3f} |"
print(row_full)

row_test = "| –†–µ–∂–∏–º —ç–∫–∑–∞–º–µ–Ω–∞         |"
for k in range(-3, 4):
    row_test += f" {stats_test[k]:>5.2f} |"
row_test += f" {within_one_test:>9.3f} |"
print(row_test)
```

# –ö–õ–ê–°–°–ò–§–ò–ö–ê–¶–ò–Ø –ù–ê –û–°–ù–û–í–ï –õ–û–ì–ò–¢-–ú–û–î–ï–õ–ò –ú–ù–û–ñ–ï–°–¢–í–ï–ù–ù–û–ì–û –í–´–ë–û–†–ê

# –û–±—É—á–µ–Ω–∏–µ –ª–æ–≥–∏—Ç-–º–æ–¥–µ–ª–∏

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix

logit_model = LogisticRegression(
    multi_class='multinomial',
    solver='lbfgs',
    max_iter=1000,
    random_state=42
)
logit_model.fit(X, y)

y_pred = logit_model.predict(X)
y_proba = logit_model.predict_proba(X)
```

# –ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã –º–æ–¥–µ–ª–∏

```python
n_classes = len(logit_model.classes_)
n_features = X.shape[1]

coeff_df = pd.DataFrame(
    logit_model.coef_,
    columns=X.columns,
    index=[f'–ö–ª–∞—Å—Å {i+1}' for i in range(n_classes)]
)
coeff_df['–ö–æ–Ω—Å—Ç–∞–Ω—Ç–∞'] = logit_model.intercept_
print("–ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –∫–ª–∞—Å—Å–∞:")
coeff_df
```

# –ö–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–æ–Ω–Ω–∞—è —Ç–∞–±–ª–∏—Ü–∞

```python
cm = confusion_matrix(y, y_pred)
percentages = np.zeros((4, 4))
for i in range(4):
    row_sum = cm[i].sum()
    for j in range(4):
        percentages[i, j] = (cm[i, j] / row_sum) * 100 if row_sum > 0 else 0

print("| –ù–∞–±–ª—é–¥–∞–µ–º—ã–µ | –ü—Ä–æ–≥–Ω–æ–∑–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è                            |")
print("|  –∑–Ω–∞—á–µ–Ω–∏—è   |   ùë¶ÃÖ = 1   |   ùë¶ÃÖ = 2   |   ùë¶ÃÖ = 3   |   ùë¶ÃÖ = 4   |")
print("|" + "-"*80 + "|")
for i in range(4):
    row = f"|   ùëìÃÖ = {i+1}    |"
    for j in range(4):
        row += f" {percentages[i, j]:>10.2f} |"
    print(row)
print("|" + "-"*80 + "|")

print("\n–¢–æ—á–Ω–æ—Å—Ç—å –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏ –ø–æ –∫–ª–∞—Å—Å–∞–º:")
for i in range(4):
    accuracy = percentages[i, i]
    print(f"  –ö–ª–∞—Å—Å {i+1}: {accuracy:.2f}%")
total_accuracy = np.trace(cm) / cm.sum() * 100
print(f"\n–û–±—â–∞—è —Ç–æ—á–Ω–æ—Å—Ç—å: {total_accuracy:.2f}%")
```

# –û—à–∏–±–∫–∏ –ø—Ä–æ–≥–Ω–æ–∑–æ–≤

```python
deviations = y_pred - y
dev_counts = {}
for k in range(-3, 4):
    dev_counts[k] = np.sum(deviations == k)

total = len(deviations)
dev_percentages = {k: (count / total) * 100 for k, count in dev_counts.items()}

exact_matches = dev_percentages[0]
within_one_class = sum(dev_percentages[k] for k in range(-1, 2))

print("| -3  |  -2   |   -1    |    0    |    1    |   2   |  3  | |Œî|‚â§1    |")
print("|" + "-"*60 + "|")
row = "|"
for k in range(-3, 4):
    row += f" {dev_percentages[k]:>6.2f} |"
row += f" {within_one_class:>8.2f} |"
print(row)
print("|" + "-"*60 + "|")

print(f"\n–î–æ–ª—è —Ç–æ—á–Ω—ã—Ö –ø—Ä–æ–≥–Ω–æ–∑–æ–≤: {exact_matches:.2f}%")
print(f"–î–æ–ª—è –ø—Ä–æ–≥–Ω–æ–∑–æ–≤ —Å –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ–º ‚â§1 –∫–ª–∞—Å—Å: {within_one_class:.2f}%")
```
